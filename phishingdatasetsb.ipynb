```python
import matplotlib.pyplot as plt

# Example data (you can replace with your own)
pie_labels = ["TCP (6)", "TCP (6)", "TCP (6)", "TCP (6)", "UDP (17)"]
pie_sizes = [6, 6, 6, 6, 17]

# Example scatter data
bwd_packet_length = [50, 80, 420, 450, 20]
source_port = [300, 700, 850, 780, 200]

# Create side-by-side plots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Pie chart
axs[0].pie(pie_sizes, labels=pie_labels, startangle=90)
axs[0].set_title("Protocol Distribution")

# Scatter plot
axs[1].scatter(bwd_packet_length, source_port)
axs[1].set_xlabel("Bwd Packet Length")
axs[1].set_ylabel("Source Port")
axs[1].set_title("Source Port vs Bwd Packet Length")

plt.tight_layout()
plt.show()
```
